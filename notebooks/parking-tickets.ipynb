```python
from tqdm import tqdm
```

```python
subframes = []
# We read in parts because this file has 10 million records in it, more than my PC can handle.
for rowstart in tqdm(range(0, 11000000, 1000000)):
    rowend = rowstart + 1000000
    rowend = min([rowend, 10626899])
    tickets = pd.read_csv("Parking_Violations_Issued_-_Fiscal_Year_2016.csv", skiprows=rowstart, nrows=rowend - rowstart)
    # We have to reset columns to terrible generics because all of the reads past the first one will use the first row
    # (of data!) as their column headers.
    tickets.columns = ['column_' + str(c) for c in list(range(len(tickets.columns)))]
    # column_2 is the state code.
    tickets = tickets[tickets['column_2'].isin(state_codes[:,1])]
    # column_13 is the police precint assigned to the ticket-issued location
    tickets_by_precinct = tickets[['column_1', 'column_2', 'column_13']].groupby(['column_13', 'column_2']).count()
    subframes.append(tickets_by_precinct)
```

```python
import geopandas as gpd
precincts = gpd.read_file("Police Precincts.geojson")
```

```python
precinct_nums = np.array([int(n) for n in np.unique(precincts['precinct'])])
```

```python
len(precinct_nums)
```

```python
precinct_state_tickets = pd.concat(subframes)
```

```python
precinct_state_tickets_dat = precinct_state_tickets.reset_index().query('column_13.isin(@precinct_nums)')
precinct_state_tickets_dat.columns = ['Precinct', 'State Code', 'Count']
```

```python
statemap = {abbr: fullname for fullname, abbr in state_codes}
```

```python
precinct_state_tickets_dat['State Name'] = precinct_state_tickets_dat['State Code'].map(statemap)
```
